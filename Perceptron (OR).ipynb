SINGLE LAYER FEEDFORWARD NEURAL NETWORK (PERCEPTRON)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.float16(np.array([0, 0, 1, 1])) #Values of Input x1
x2 = np.float16(np.array([0, 1, 0, 1])) #Values of Input x2
y_d = np.float16(np.array([0, 1, 1, 1]))
```

```python
lr = np.float16(0.1)
theta = np.float16(0.3)   #Bias
w1 = np.float16(0.4)
w2 = np.float16(-0.8)
no_of_epochs = 20    #For training
```

```python
no_of_iterations =  len(x1) * no_of_epochs
print("Number of iterations: %d" %(no_of_iterations))
```

```python
error = []
we_1 = []
we_2 = []
for j in range(no_of_epochs):
  for i in range(len(x1)):
    sum = (x1[i]*w1)+(x2[i]*w2)       #computes the net weighted input
    if sum >= theta:                  #compares with the threshold (Bias)
      y_act = np.float16(1.)
    else:
      y_act = np.float16(0.)
    err = y_d[i] - y_act              #error for a single computation
    error.append(err)
    delta_w1 = lr * x1[i] * err
    delta_w2 = lr * x2[i] * err
    w1 = w1 + delta_w1
    w2 = w2 + delta_w2
    we_1.append(w1)
    we_2.append(w2)
```

```python
#print("Length of error values: %d" %(len(error)))

print("Error: ", error)
print("Weight w1: ", we_1)
print("Weight w2: ", we_2)
```

```python
#The Model Parameters (learnt after training)
print("Weight, w1 = %f" %w1)
print("Weight, w2 = %f" %w2)
```

```python
#Visualize error variation with respect to number of epochs
epochs = np.linspace(1,no_of_epochs,no_of_iterations)
plt.plot(epochs, error)
plt.title("Error vs. Epochs")
plt.xlabel("Number of epochs")
plt.ylabel("Error")
plt.show()
```

```python
#Visualize variations in weights with respect to number of epochs
epochs = np.linspace(1,no_of_epochs,no_of_iterations)
plt.plot(epochs, we_1)
plt.title("Variations in Weight, w1 vs. Epochs")
plt.xlabel("Number of epochs")
plt.ylabel("Weight, w1")
plt.show()
plt.plot(epochs, we_2)
plt.title("Variations in Weight, w2 vs. Epochs")
plt.xlabel("Number of epochs")
plt.ylabel("Weight, w2")
plt.show()
```
